```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("Opal_Tecopa_near_gem.jpg",0) 
img = cv2.medianBlur(img,5)
print(img.min())
```

```python
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,3,20,
                            param1=150,param2=50,minRadius=0,maxRadius=30)
x = circles
x[0][:,2]# This can give us the radius of the particles
```

```python
circles = np.uint16(np.around(circles))
print(circles)
```

```python
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(img,(i[0],i[1]),2,(0,0,255),3)
plt.imshow(img)
```
