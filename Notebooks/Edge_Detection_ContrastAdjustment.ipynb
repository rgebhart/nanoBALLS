```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('SEM_Images/R5 K3Me3 SiO2_3-3.tif')
img = np.array(img, dtype=np.uint8)
```

```python
kernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
canny = cv2.Canny(opening,100,150,3,L2gradient=True)
#fig, ax = plt.subplots(3,1,figsize=(30,30))
#ax[0].imshow(opening)
#ax[1].imshow(canny)
#ax[2].imshow(img)


#try to play with contrast by spreading out the histrogram with a factor alpha

alpha = 1.6 # Simple contrast control
beta = 0    # Simple brightness control
img2 = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
kernel = np.ones((5,5), np.uint8)
opening2 = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)
canny2 = cv2.Canny(opening2,100,150,3,L2gradient=True)

#ax[0].imshow(opening2)
#ax[1].imshow(canny2)
#ax[2].imshow(img2)

fig, ax = plt.subplots(2,2,figsize=(30,30))
ax[0][0].imshow(img)
ax[0][1].imshow(canny)
ax[1][0].imshow(img2)
ax[1][1].imshow(canny2)

fig.subplots_adjust(left= None, bottom=0.5, right=None, top=1, wspace=None, hspace=None)
```
